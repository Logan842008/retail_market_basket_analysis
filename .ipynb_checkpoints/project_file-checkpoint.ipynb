```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
products = pd.read_csv('data/products.csv')
transactions = pd.read_csv('data/transactions.csv')
stores = pd.read_csv('data/stores.csv')
```

```python
# Data Cleaning & Preprocessing
transactions['date'] = pd.to_datetime(transactions['date'])  # Convert date column to datetime
transactions['date']
```

```python
products.fillna('Unknown', inplace=True)
stores.fillna('Unknown', inplace=True)
transactions.fillna(0, inplace=True)
```

```python
# Join datasets for enrichment
# Feature Engineering
transactions['total_price'] = transactions['quantity'] * transactions['unit_price']  # Calculate total spend per product
transactions['day_of_week'] = transactions['date'].dt.day_name()  # Extract day of the week
transactions['hour_of_day'] = transactions['date'].dt.hour  # Extract hour of the day

transactions = transactions.merge(products, left_on='category', right_on='category', how='left')
transactions = transactions.merge(stores, on='store_id', how='left')
transactions
```

```python
store_metrics = transactions.groupby('store_id').agg(
    total_revenue=('total_price', 'sum'),
    unique_transactions=('transaction_id', 'nunique'),
    total_items=('quantity', 'sum')
).reset_index()
```

```python
store_metrics['average_transaction_value'] = store_metrics['total_revenue'] / store_metrics['unique_transactions']
store_metrics['average_transaction_value']
```

```python
# Save processed data
transactions.to_csv('data/processed_transactions.csv', index=False)
store_metrics.to_csv('data/store_metrics.csv', index=False)
```

```python
# Display summary for validation
print("Data cleaned and processed. Summary:")
store_metrics.head()
```

```python
# Save enriched transactions for further analysis
transactions.head()
```

```python
# Association Rule Mining
basket = (transactions.groupby(['transaction_id', 'product_id'])
          ['quantity'].sum().unstack().fillna(0).map(lambda x: 1 if x > 0 else 0))
```

# Logan

```python
frequent_itemsets = apriori(basket, min_support=0.01, use_colnames=True, low_memory=True)
rules = association_rules(frequent_itemsets, num_itemsets=len(frequent_itemsets), metric="lift", min_threshold=1.0)
rules = rules.sort_values('lift', ascending=False)
```

```python
rules.to_csv('data/association_rules.csv', index=False)
```

```python
high_freq_categories = frequent_itemsets.sort_values('support', ascending=False).head(5)['itemsets']
print("High-frequency product categories:", high_freq_categories.tolist())
```

```python
seasonal_promo = transactions.groupby(transactions['date'].dt.month)['total_price'].sum()
promo_calendar = seasonal_promo.reset_index()
promo_calendar.columns = ['Month', 'Revenue']
```

```python
promo_calendar.to_csv('data/promo_calendar.csv', index=False)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Month', y='Revenue', data=promo_calendar)
plt.title('Seasonal Revenue Trends')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('figures/seasonal_trends.png')
```

```python
current_average_transaction_value = 45
new_average_transaction_value = store_metrics['average_transaction_value'].mean()
increase_percentage = ((new_average_transaction_value - current_average_transaction_value) / current_average_transaction_value) * 100
```

```python
print(f"Current Average Transaction Value: ${current_average_transaction_value}")
print(f"New Average Transaction Value: ${new_average_transaction_value:.2f}")
print(f"Increase in Transaction Value: {increase_percentage:.2f}%")
```
